```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
papers_df = pd.read_csv("papers/cl_cv_papers_errors.csv")
# group the papers based on year and month and compute both the mean and the variance for num_errors
agg_df = (
    papers_df.groupby(["year", "month"])
    .agg({"num_errors": ["mean", "var"]})
    .reset_index()
)
# make the columns just year, month, error_mean, error_var
agg_df.columns = ["year", "month", "error_mean", "error_var"]
agg_df.dropna(inplace=True)
agg_df.set_index(["year", "month"], inplace=True)
agg_df.sort_index(inplace=True)
# make a date column using the year and month columns
agg_df["date"] = pd.to_datetime(
    agg_df.index.to_flat_index().map(lambda x: f"{x[0]}-{x[1]}")
)
# draw the plot, one for the mean and one for the variance using plotly
fig = px.line(
    agg_df, x="date", y="error_mean", title="Mean number of errors per paper per month"
)
fig.show()

fig = px.line(
    agg_df,
    x="date",
    y="error_var",
    title="Variance of number of errors per paper per month",
)
fig.show()
```
